```python

''''
2.why threading module used? write the use of the following functions-
activeCount()
currentThread()
enumerate()

ANs-The threading module in Python is used to create and manage threads, which are independent execution contexts that can run concurrently within the same process. Threads can be used to perform parallel or asynchronous tasks, which can help improve the performance of certain types of applications.

Here are the explanations for the following functions in the threading module:

activeCount(): This function returns the number of Thread objects that are currently active in the program. An active thread is one that has been started but has not yet finished executing. The function can be used to monitor the progress of concurrent tasks and to ensure that all threads have completed before the program exits.

currentThread(): This function returns a reference to the current Thread object that is executing. The Thread object represents the current thread and provides access to various attributes and methods that can be used to control and monitor the thread's behavior.

enumerate(): This function returns a list of all active Thread objects in the program. Each element in the list is a reference to a Thread object, which can be used to access its attributes and methods. The function can be used to inspect and monitor the state of all active threads in the program.


'''

import threading
import time

# Define a function to be executed by a thread
def my_task():
    print(f"Starting task on thread {threading.current_thread().name}")
    time.sleep(1)
    print(f"Finishing task on thread {threading.current_thread().name}")

# Create a list of thread objects and start them
threads = []
for i in range(5):
    t = threading.Thread(target=my_task)
    threads.append(t)
    t.start()

# Wait for all threads to complete
for t in threads:
    t.join()

# Print the number of active threads and a list of all active threads
print(f"Number of active threads: {threading.active_count()}")
print(f"List of active threads: {threading.enumerate()}")

```

```python

'''Q4.write python program to create two threads. Thread one must print the list of squares and thread two must print the list of cubes
'''
import threading
def print_squares():
    for i in range(1, 11):
        print(f"{i} squared is {i**2}")

# Define a function to print the cubes of numbers
def print_cubes():
    for i in range(1, 11):
        print(f"{i} cubed is {i**3}")

# Create two threads
thread1 = threading.Thread(target=print_squares)
thread2 = threading.Thread(target=print_cubes)

# Start both threads
thread1.start()
thread2.start()

# Wait for both threads to complete
thread1.join()
thread2.join()
```

Q5. State advantages and disadvantages of multithreading?

Ans--Multithreading is a programming technique where multiple threads of execution are created within a single process. In Python, multithreading is commonly used for concurrent and parallel programming. Below are some advantages and disadvantages of multithreading in Python:

Advantages:

Improved Performance: Multithreading allows multiple threads to run concurrently, which can improve performance by utilizing multiple CPU cores and reducing idle time.
Asynchronous Execution: Threads can be designed to run independently and asynchronously, allowing for tasks to be executed in the background without blocking the main thread.
Responsiveness: Multithreading can help make user interfaces more responsive by offloading long-running tasks to a separate thread, allowing the main thread to remain responsive to user input.
Resource Sharing: Multithreading can facilitate efficient sharing of resources such as memory and file handles, as threads within the same process can access and modify the same resources.
Simplified Code: Multithreading can simplify code that would otherwise be complex and difficult to manage, especially in cases where multiple tasks need to be performed simultaneously.
Disadvantages:

Synchronization: Multithreading requires proper synchronization of shared resources to avoid race conditions and other concurrency-related issues, which can be difficult to manage and debug.
Overhead: Multithreading introduces overhead due to the creation, management, and synchronization of threads, which can lead to decreased performance if the overhead outweighs the benefits of parallel execution.
Deadlocks: Multithreading can lead to deadlocks if multiple threads attempt to acquire shared resources in a circular manner, preventing any of the threads from progressing.
Limited Global Interpreter Lock (GIL): In Python, the GIL limits the parallel execution of multiple threads, which can limit the performance gains from multithreading.
Debugging: Multithreaded programs can be difficult to debug due to the unpredictable nature of thread scheduling and the possibility of race conditions and deadlocks.
Overall, multithreading in Python can be a powerful tool for improving performance and managing complex tasks, but it requires careful consideration of its advantages and disadvantages and proper implementation to avoid common pitfalls.

Deadlocks and race conditions are two types of synchronization problems that can occur in concurrent programming, including in Python.

A deadlock occurs when two or more processes or threads are blocked and unable to proceed because they are waiting for each other to release resources. For example, if two threads are each holding a resource that the other thread needs, they can become deadlocked, as neither thread can proceed without first obtaining the resource held by the other thread. Deadlocks can occur in Python when multiple threads or processes are accessing shared resources like files, databases, or network connections.

A race condition occurs when the behavior of a program depends on the timing of how multiple threads or processes execute, and the outcome of the program is not deterministic or predictable. For example, if two threads are incrementing a counter variable, they can end up overwriting each other's updates, leading to incorrect results. Race conditions can occur in Python when multiple threads or processes are accessing shared data structures like lists or dictionaries.

To avoid deadlocks and race conditions in Python, it is important to use appropriate synchronization mechanisms like locks, semaphores, or queues. These mechanisms can help ensure that only one thread or process accesses a shared resource at a time, preventing deadlocks and race conditions from occurring. It is also important to design programs with concurrency in mind, taking care to identify and manage shared resources to minimize the likelihood of synchronization problems.
